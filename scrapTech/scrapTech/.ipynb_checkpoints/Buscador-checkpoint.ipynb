# Gestor de tablas

### Llegados a este punto y una vez realizadas las extracciones crearemos una tabla acumulada para cada página web.

Para crear una tabla acumulada utilizaremos la última tabla extraída y la añadiremos a una tabla actual. Después utilizamos tabla-1 para ir añadiendo los componentes actuales 


```python
import pandas as pd
import Loader
#Cargamos las dos ultimas tablas
actual = './pccomponentes/08-46_14-05_pccomponentes.json'
ant = './pccomponentes/21-28_13-05_pccomponentes.json'

#Carga en nuestra app
# l = Loader()
# #Cargador de datasets 
# actual = Loader.cargar_datos(actual)

actual = pd.read_json(actual, lines=True)
ant = pd.read_json(ant, lines=True)
```

```python
print("Elementos en actual: " , actual.count())
print("Elementos en ant: ", ant.count())
```

#### En PCComponentes, los elementos que ya no estén es porque no aparecen en la web, por tanto simplemente a no disponibles, ya que si entramos a su enlace, seguirán apareciendo.

Funciones:

merge how=outer -> une las filas de uno a otro
merge -> une añadiendo columnas

```python
aunEstan = actual.merge(ant, how='outer')
aunEstan.count()
```

#### Buscamos los elementos de anterior que no están en actual y los de actual que no están en anterior

Componentes que no están en actual pero sí en anterior
Dos casos:

    wipoid -> los que ya no aparecen en la tabla se eliminan. (Wipoid gestiona la disponibilidad con Disponible NO disponible)
    pccomponentes -> los que ya no aparecen en la tabla se ponen a No Disponibles (PCcomponentes no muestra lo que no hay activo pero la web sigue existiendo, así que se pone a no disponible)

```python
#Df con los componentes de anterior que ya no están en actual. Componentes No_Disponibles
aniadidos = ant[(~ant.id_item.isin(actual.id_item))]
aniadidos.count()
```

Componentes de actual que no están en anterior es porque han sido añadidos o ha vuelto a haber stock
Dos casos:

    wipoid -> Puede darse que haya disonibilidad (poner a DISPONIBLE TRUE) o puede ser que sean nuevos (añadir fila)
    pccomponentes -> Puede ser que sean nuevos (comparar con id_item) o que se hayan añadido como componentes nuevos (añadir fila)



```python
#Df con los componentes de actual que no están en anterior. Componentes aniadidos
eliminados = actual[(~actual.id_item.isin(ant.id_item))]
eliminados.count()
```
