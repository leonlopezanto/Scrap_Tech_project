#### Buscamos los elementos comunes de anterior y actual para ver si el precio ha variado



Si el precio de anterior > actual -> BAJADA DE PRECIO, FIESTA\
Si el precio de anterio  < actual -> SUBIDA DE PRECIO, MIERDA\
Si el precio de anterior = actual -> SE MANTIENE Y NO SE NOTIFICA

Según ocurra esto, el campo, precio_actualizado = baja/sube/mantiene

Obtenemos los elementos comunes entre anterior y actual y reseteamos el indice para evitar problemas

```python
#ant = ant[(ant.id_item.isin(actual.id_item))][['id_item','precio']].reset_index()
#act = actual[(actual.id_item.isin(ant.id_item))][['id_item','precio']].reset_index()
an = ant[(ant.id_item.isin(actual.id_item))].reset_index()
ac = actual[(actual.id_item.isin(ant.id_item))].reset_index()
```

Nos quedamos con los componentes que se repiten en ambos y añadimos la nueva columna a actual

```python
import numpy as np
# Lista de condiciones
conditions = [
    (an['precio'] < ac['precio']),
    (an['precio'] > ac['precio']),
    (an['precio'] == ac['precio'])
    ]
# lista de valores a asignar a las condiciones
values = ['sube', 'baja', 'mantiene']

#Utilizamos el metodo select de Numpy
ac['precio_upd'] = np.select(conditions, values)
ac
```

Actualizamos la fecha de actualizacion si ha habido un cambio

```python
#ac.loc['update'] = ac['precio_upd'].apply(lambda x: date if x != 'mantiene' else ac.loc['update'])
ac['update'].loc[(ac['precio_upd'] == 'sube') | (ac['precio_upd'] == 'baja')] = get_fecha()
ac
```

Ya tenemos la tabla con precios actualizados de los elementos que se repiten. Ahora necesitamos la tabla con elementos que NO se repiten.

```python
def mantenidos(ant, actual):
    
    an = ant[(ant.id_item.isin(actual.id_item))].reset_index()
    ac = actual[(actual.id_item.isin(ant.id_item))].reset_index()
    conditions = [
        (an['precio'] < ac['precio']),
        (an['precio'] > ac['precio']),
        (an['precio'] == ac['precio'])
    ]
    # lista de valores a asignar a las condiciones
    values = ['sube', 'baja', 'mantiene']

    #Utilizamos el metodo select de Numpy
    ac['precio_upd'] = np.select(conditions, values)
    
    ac['update'].loc[(ac['precio_upd'] == 'sube') | (ac['precio_upd'] == 'baja')] = get_fecha()
    
    del ac['index']

    return ac
```

```python
#Cargamos dos tablas
a = 'actual_peq.json'
b = 'anterior_peq.json'

actual2 = pd.read_json(a, lines=True)
ant2 = pd.read_json(b, lines=True)

mantenidos = mantenidos(maestraPC, actual2)
maestraPC
```
