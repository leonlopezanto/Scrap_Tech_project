## La maravillosa tabla maestra contiene TODA LA INFORMACIÓN NECESARIA que se debe presentar a la interfaz web.

La tabla maestra se nutre de tablas submaestras para cada araña. En principio para PcComponentes habrá una submaestra y para Wipoid habrá otra. Estas tablas submaestras se nutrirán y actualizarán con información del último scrap realizado, de manera que se actualizarán constantemente tras cada etapa de scrap. 

Para crear una tabla acumulada utilizaremos la última tabla extraída y la añadiremos a una tabla actual. Después utilizamos tabla-1 para ir añadiendo los componentes actuales 


```python
import pandas as pd

import sys
sys.path.append('../../')
import Loader
from datetime import datetime

def get_fecha(tipo = 'normal'):
    return datetime.now().strftime("%d-%b-%Y (%H:%M:%S)") if(tipo == 'normal') else datetime.now().strftime("%H-%M_%d-%m")


#Cargamos las dos ultimas tablas
actual = 'actual.json'
ant = 'anterior.json'

#Carga en nuestra app
# l = Loader()
# #Cargador de datasets 
# actual = Loader.cargar_datos(actual)

actual = pd.read_json(actual, lines=True)
ant = pd.read_json(ant, lines=True)
```

### PCComponentes
#### Buscamos los elementos de anterior que no están en actual 

Componentes que no están en actual pero sí en anterior

    pccomponentes -> los que ya no aparecen en la tabla se ponen a No Disponibles (PCcomponentes no muestra lo que no hay activo pero la web sigue existiendo, así que se pone a no disponible)

```python
#Df con los componentes de anterior que ya no están en actual. Componentes No_Disponibles
eliminados = ant[(~ant.id_item.isin(actual.id_item))]

print('Mostrando elementos de anterior que han sido descartados de la web')
eliminados
```

Los elementos que han sido descartados de la web es que han dejado de estar disponibles. disponible=False, update=fecha_actual


```python
#eliminados.is_copy = None
#Elimina la excepcion. NO nos importa
pd.set_option('mode.chained_assignment', None)
#actualizamos campo disponible
eliminados['disponible'] = False
#actualizamos campo update
eliminados['update'] = get_fecha()

#Unimos los archivos que no estan disponibles a la tabla maestra de pccomponentes
maestraPC = nuevo.merge(eliminados, how='outer')
#variable maestraPC con elementos no disponibles y disponibles actualizados y update actualizado
#maestraPC
```

#### Buscamos los elementos comunes de anterior y actual para ver si el precio ha variado



Si el precio de anterior > actual -> BAJADA DE PRECIO, FIESTA\
Si el precio de anterio  < actual -> SUBIDA DE PRECIO, MIERDA\
Si el precio de anterior = actual -> SE MANTIENE Y NO SE NOTIFICA

Según ocurra esto, el campo, precio_actualizado = baja/sube/mantiene

```python
an = ant[(ant.id_item.isin(actual.id_item))][['id_item','precio']].reset_index()
ac = actual[(actual.id_item.isin(ant.id_item))][['id_item','precio']].reset_index()

print(an)
print(ac)
```

```python
import numpy as np
# create a list of our conditions
conditions = [
    (an['precio'] < ac['precio']),
    (an['precio'] > ac['precio']),
    (an['precio'] == ac['precio'])
    ]
# create a list of the values we want to assign for each condition
values = ['sube', 'baja', 'mantiene']

ac['precio_upd'] = np.select(conditions, values)

ac
```
